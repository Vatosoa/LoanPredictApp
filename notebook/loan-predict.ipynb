<h1>CrÃ©ation du modÃ¨le</h1>

```python
df_cat=df[var_cat]
df_cat
```

```python
df_cat = pd.get_dummies(df_cat,drop_first=True) #creer deux colonne pour les variables dans la colonne
df_cat
```

```python
data = {
    'Credit_History': [1.0, 0.0, 1.0, 1.0, 0.0],
    'Gender': ['Male', 'Female', 'Female', 'Male', 'Female'],
    'Married': ['Yes', 'No', 'Yes', 'No', 'Yes'],
    'Dependents': ['0', '1', '2', '3+', '1'],
    'Education': ['Graduate', 'Not Graduate', 'Graduate', 'Graduate', 'Not Graduate'],
    'Self_Employed': ['No', 'Yes', 'No', 'Yes', 'No'],
    'Property_Area': ['Urban', 'Rural', 'Semiurban', 'Urban', 'Semiurban'],
    'Loan_Status': ['Y', 'N', 'Y', 'N', 'Y']
}

df_data = pd.DataFrame(data)

# Utiliser pd.get_dummies pour convertir les colonnes catÃ©goriques en variables binaires
df_cate = pd.get_dummies(df_data, drop_first=True)

# Convertir les valeurs boolÃ©ennes en entiers (0 et 1)
df_cate = df_cat.astype(int)

# Afficher le DataFrame transformÃ©
print("DataFrame aprÃ¨s transformation en variables binaires (0 et 1) :")
df_cate
```

```python
df.columns
```

```python
var_num
df[var_num]
```

```python
df_num=df[var_num]
```

```python
df_encoded = pd.concat([df_cate, df_num], axis=1)
df_encoded
```

```python
y=df_encoded['Loan_Status_Y']
y
```

```python
X = df_encoded.drop('Loan_Status_Y', axis=1)
X
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=6)
```

```python
# instancier le modÃ¨le
clf = LogisticRegression()
# entrainer le modÃ¨le
clf.fit(x_train, y_train)
```

```python
# tester le modÃ¨le
pred = clf.predict(x_test)
```

```python
pred
```

```python
y_test
```

```python
accuracy_score(y_test, pred)
```
